### figure对象的定义

#### 方法一

```python
import matplotlib.pyplot as plt
import numpy as np
fig  = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax3.plot(np.random.randn(50),"k--")
```

#### 方法二

```python
fig, axes = plt.subplots(2,3) # 创建figure对象的同时创建2*3=6个子图 注意axes是数组
axes[1][1].plot(np.random.randn(20)) # 访问第1行1列的子图
plt.subplots_adjust(wspace=1,hspace=1) # 设置每个子图之间的高度间隙和宽度间隙, 更多参数见文档
```

### 刻度、标签和图例

```python
fig, axes = plt.subplots(2,2)
axes[0][0].plot(np.random.randn(20))
plt.subplots_adjust(wspace=0.5, hspace=1)
axes[0][0].set_xlim(0, 100)              # 设置x轴从0到100
axes[0][0].set_xticks(range(0,100,10))   # 设置x轴的刻度是0 10 20 30...100
axes[0][0].set_title(u"随机图")          # 设置某个图的名字 有时候中文打不出来
axes[0][0].set_xlabel("x轴")             # 设置x轴的标签
```

#### 图例

```python
fig, axes = plt.subplots(1,1)
for i in range(0,2):
    axes.plot(np.random.randn(20), label=i)
axes.legend(loc="best")                    # 画出图例 best参数能让图例自动找到最合适的位置出现
axes.text(15,2,"hi")                       # 设置在图的（x,y）位置上添加一段文字
```

### Pandas 画图

```python
import pandas as pd
data = pd.DataFrame()
data["x1"]=range(1,100)
data["x2"]=np.random.randn(99)

data["x2"].plot()              # 直接画出x2的数据
```

```python
data.plot(xlim=(0,10), xticks=(range(0,100,10)))   # 直接画出所有数据的图 跟matplot一样的设置
```

```python
data["x1"].plot(kind='barh')   # 设定画出水平条形图
```

```python
# 画出密度函数和直方图
data["x2"].plot(kind='kde')   
data["x2"].hist(bins=30,alpha=.5)
```
