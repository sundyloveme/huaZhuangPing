```python
import matplotlib.pyplot as plt
import pandas as pd
# 加载化妆品和置装费的数据，并且清洗掉不要的数据
dataSet = pd.read_csv("someData.csv")
dataSet["huaZhuangPingCost"]=dataSet["化妆品费用"]
dataSet["yiFuCost"]=dataSet["置装费"]
dataSet.dropna(axis=1)# 清理掉列值为空的数据
dataSet = dataSet[["huaZhuangPingCost","yiFuCost"]]

huaZhuangPingCostAvg = dataSet["huaZhuangPingCost"].mean() # 求该列的平均值
yiFuCostAvg = dataSet["yiFuCost"].mean()
dataSet
```

```python
print(huaZhuangPingCostAvg,yiFuCostAvg)
```

```python
dataSet["x-x^"] = dataSet["huaZhuangPingCost"] - huaZhuangPingCostAvg
dataSet["y-y^"] = dataSet["yiFuCost"] - yiFuCostAvg
dataSet["(x-x^)^2"] = dataSet["x-x^"] **2
dataSet["(y-y^)^2"] = dataSet["y-y^"] **2
dataSet
```

```python
# 让我们画出散点图
fig, ax = plt.subplots()
ax.scatter(dataSet["huaZhuangPingCost"], dataSet["yiFuCost"])
ax.set_xlabel("huaZhuangPingCost")
ax.set_ylabel("yiFuCost")
```

# Find
不难看出买化妆品开销大的人买衣服开销一样大
